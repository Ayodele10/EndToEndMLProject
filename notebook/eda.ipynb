CREATING A VARIABLE FOR TOTAL AND AVERAGE SCORES

```python
stud_data['total score'] = stud_data['math score'] + stud_data['reading score'] + stud_data['writing score'] # Creating a new column 'total score'

stud_data['average score'] = stud_data['total score'] / 3 # Creating a new column 'average score'
stud_data['average score'] = stud_data['average score'].round(2) # Rounding off the average score to 2 decimal places


stud_data.head() # Displaying the first five rows of the dataset
```

DATA VISUALISATION

```python
plt.figure(figsize=(10,6))
sns.histplot(stud_data['average score'], bins=30, kde=True)
plt.title('Distribution of Average Scores')
plt.xlabel('Average Score')
plt.ylabel('Frequency')
```
